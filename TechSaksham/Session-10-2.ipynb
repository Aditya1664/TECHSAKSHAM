```python
#creating multiple subplots in one figure
x1_axis=[2,4,6,8]
y1_axis=[1,10,100,1000]

x2_axis=[1,3,5,7]
y2_axis=[100,110,120,130]
```

```python
#plt.subplot(no_of-rows,no_of_columns,current_plot_number)
plt.subplot(1,2,1)
plt.plot(x1_axis,y1_axis,'r')

plt.subplot(1,2,2)
plt.plot(x2_axis,y2_axis,'g')
plt.tight_layout()
```

```python
plt.subplot(2,1,1)
plt.title('Exponential')
plt.plot(x1_axis,y1_axis,'r')

plt.subplot(2,1,2)
plt.title('Linear')
plt.plot(x2_axis,y2_axis,'g')
plt.tight_layout()
```
