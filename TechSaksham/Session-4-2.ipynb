```python
password = input("Enter your password: ")
strength = 0
if len(password) >= 8:
    strength += 1
else:
    print("Weak: Password should be at least 8 characters long.")
has_upper = False
has_lower = False
has_digit = False
has_special = False
special_chars = "!@#$%^&*(),.?\":{}|<>"
for char in password:
    if char.isupper():
        has_upper = True
    if char.islower():
        has_lower = True
    if char.isdigit():
        has_digit = True
    if char in special_chars:
        has_special = True
if has_upper:
    strength += 1
else:
    print("Weak: Password should contain at least one uppercase letter.")
if has_lower:
    strength += 1
else:
    print("Weak: Password should contain at least one lowercase letter.")
if has_digit:
    strength += 1
else:
    print("Weak: Password should contain at least one number.")
if has_special:
    strength += 1
else:
    print("Weak: Password should contain at least one special character.")
if strength == 5:
    print("Strong: Your password is secure!")
```

```python
a, b, c = map(int, input("Enter three numbers: ").split())
if a == b == c:
    print("Sum: 0")
elif a == b:
    print("Sum:", c)
elif b == c:
    print("Sum:", a)
elif a == c:
    print("Sum:", b)
else:
    print("Sum:", a + b + c)
```

```python
import random
letters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
numbers = "0123456789"
special_chars = "!@#$%^&*()_+-=[]{}|;:,.<>?/"
length = int(input("Enter the desired password length: "))
password = [
    random.choice(letters), 
    random.choice(numbers), 
    random.choice(special_chars)
]
all_chars = letters + numbers + special_chars
password += random.choices(all_chars, k=length - 3)
random_password = "".join(random.sample(password, len(password)))
print("Generated Password:", random_password)
```

```python
st = "Walchand Institute of Technology,Solapur,Maharashtra"
alphabet_count = {}
i=0
while i<len(st): 
    if st[i].isalpha():  
        char = st[i].lower() 
        if char in alphabet_count:
            alphabet_count[char] += 1
        else:
            alphabet_count[char] = 1
    i+=1
else:
    print("Alphabet Counts:", alphabet_count)
```

```python
print('''\nReady to play 'Rock','Paper' and 'Scissor'?
Lets Play!!!!''')
import random
while True:
    rock=(''' Rock
         _______
     ---'   ____)
          (_____)
          (_____)
          (____)
    ---.__(___)
    ''')

    paper=(''' Paper
         _______
    ---'    ____)____
               ______)
              _______)
             _______)
    ---.__________)
    ''')

    scissor=(''' Scissor
        _______
    ---'   ____)____
              ______)
           __________)
          (____)
    ---.__(___)
    ''')
    li=[rock,paper,scissor]
    comp=random.randint(0,2)
    print('''\nEnter a number from below:
    0 for Rock,
    1 for Paper,
    2 for Scissor.''')
    user=int(input("Enter a number from 0 to 2-"))
    if user>2 or user<0: 
        print("Wrong choice. You lose.")
        break

    else:
        
        print(f"\nComputer choice is:{comp}")
        print(li[comp])
    
        print(f"User choice is:{user}")
        print(li[user])
    
        if comp==user:
            print("Its a draw.")

        elif comp==0 and user==1:
            print("User wins 🎉.")

        elif comp==0 and user==2:
            print("Computer wins ☠.")
    
        elif comp==1 and user==0:
            print("Computer wins ☠.")

        elif comp==1 and user==2:
            print("User wins 🎉.")

        elif comp==2 and user==0:
            print("User wins 🎉.")

        elif comp==2 and user==1:
            print("Computer wins ☠.")

    print("\nDO you wan tto play again?")
    a=input("Yes/No:").lower()
    if a=='no':
        break
```
